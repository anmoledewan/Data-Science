# Data Transformation

1. As most models cannot directly take in string-type variables, they need to be converted to number base form. Also for the models to differentiate between categorical and numerical data formats, ideally for nominal scale, a dummy variable is needed

2. For ordinal scale, a proper numerical conversion is needed. 

3. As Education alreast has a ordinal scale develop, we will drop education

## Column Dropping

1. As Education alreast has a ordinal scale develop, we will drop education
2. Also fnlwgt is unknown so explaining that also would be a problem

```python
df.drop('education',axis=1,inplace=True)
df.drop('fnlwgt',axis=1,inplace=True)


df_test.drop('education',axis=1,inplace=True)
df_test.drop('fnlwgt',axis=1,inplace=True)
```

## Column combination

1. Capital gain and Capital loss means the same and only have opposite signs. They are also generally related to profits and loss.

2.  A new column needs to be made for the same consisting of both values

```python
df['gainloss']=df['capital.gain']-df['capital.loss']
df

numeric_describe=df.describe().T #To calculate mean, std-dev, min and quartiles
numeric_describe
```

```python
df['gainloss'].hist(bins=5)
```

## Model 1 - Logistic Model based on 85 features

```python
train_copy = df.copy()
X_Train = train_copy.drop('income', axis =1)
Y_Train = train_copy['income']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_Train, Y_Train, 
                                                    test_size=0.33, 
                                                    random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=0)
model.fit(X_train, y_train)
```

## Testing/Evaluation of the model

```python
predicted_classes = model.predict(X_test)
```

```python

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report 
from sklearn.metrics import roc_auc_score,roc_curve,f1_score 
from sklearn.metrics import accuracy_score 
confusion_matrix(y_test,predicted_classes)
```

```python
accuracy_score(predicted_classes, y_test)
y_pred=predicted_classes
```

```python

from matplotlib import pyplot as plt
import seaborn as sns
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = "Oranges");
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Confusion Matrix - score:'+str(accuracy_score(y_test,y_pred))
plt.title(all_sample_title, size = 15);
plt.show()
print(classification_report(y_test,y_pred))
```
